```python
print('Hello! Welcome to the ticket aggregator for Moscow - New York flights.')
print('')  # Visual indent.

# Input ticket price.
# Data is not transformed to int instantly to avoid letters and other symbols.
stopper = 0
while stopper == 0:
    price = input('Please enter the desired ticket price: ')
    for i in list(price):
        if i not in list('1234567890'):
            
            # Spaces included near "!" for better visual perception.
            print('   !   Error: You entered an incorrect price, try again!')
            stopper = 0
            break
        else:
            stopper = 1
            continue

# Using a prompt.
if int(price) > 250:
    print('Your ticket will cost more than 250$ ')
elif 200 <= int(price) <= 250:
    print('Your ticket will cost between 200$ and 250$')
else:
    print('Your ticket will cost less than 200$')
print('')  # Visual indent.

# Input number of transfers.
stopper = 0
while stopper == 0:
    transfers = input('Print the suitable number of transfers: ')
    for i in list(transfers):
        if i not in list('1234567890'):
            
            # Spaces included near "!" for better visual perception.
            print('   !   Error: You entered an incorrect number, try again!')
            stopper = 0
            break
        else:
            stopper = 1
            continue

# Using a prompt.
if int(transfers) > 2:
    transfers = '3'  # Avoid big numbers of transfers.
    print('You chose a flight with 3 or more transfers.')
elif int(transfers) == (0 or 2):
    print('You chose a flight with ' + transfers + ' transfers.')
else:
    print('You chose a flight with ' + transfers + ' transfer.')
print('')  # Visual indent.

# Input refund + using a prompt.
stopper = 0
while stopper == 0:
    refund = input('Would you like to include refund in your tickets? ' 
                   'Please enter Y or N: ')
    if refund not in ('Y', 'y', 'N', 'n'):
        
        # Spaces included near "!" for better visual perception.
        print('   !   Error: You entered an incorrect answer, try again!')
    elif refund in ('Y', 'y'):
        print('Refund will be included in your ticket.')
        stopper = 1
        refund = 'included'
        break
    else:
        print('Refund will not be included in your ticket.')
        stopper = 1
        refund = 'not included'
        break
print('')  # Visual indent.

# Input luggage  + using a prompt.
stopper = 0
while stopper == 0:
    luggage = input('Would you like to include luggage in your tickets? '
                    'Please enter Y or N: ')
    if luggage not in ('Y', 'y', 'N', 'n'):
        
        # Spaces included near "!" for better visual perception.
        print('   !   Error: You entered an incorrect answer, try again!')
    elif luggage in ('Y', 'y'):
        print('Luggage will be included in your ticket.')
        stopper = 1
        luggage = 'included'
        break
    else:
        print('Luggage will not be included in your ticket.')
        stopper = 1
        luggage = 'not included'
        break
print('')  # Visual indent.

# Making the offer.
if (
    int(price) < 200 and 
    int(transfers) in (0, 1) and 
    refund == 'included' and 
    luggage == 'included'
    ):
    print(
    'The offer for Moscow - New York flight:\n'
    'Ticket category - The best!\n'
    # Used fixed price instead of "less than 200" for realism.
    'Ticket price - 180$\n' 
    'Number of transfers - ' + transfers + '\n'
    'Refund - ' + refund + '\n'
    'Luggage - ' + luggage + '\n'
    )
elif (
    200 <= int(price) <= 250 and 
    int(transfers) in (0, 1, 2)
    ):
    print(
    'The offer for Moscow - New York flight:\n'
    'Ticket category - Good enough!\n'
    # Used fixed price instead of interval for realism.
    'Ticket price - 230$\n'
    'Number of transfers - ' + transfers + '\n'
    'Refund - ' + refund + '\n'
    'Luggage - ' + luggage + '\n'
    )
elif (
    int(price) >= 250 and 
    int(transfers) >= 3
    ):
    print(
    'The offer for Moscow - New York flight:\n'
    'Ticket category - Bad :(\n'
    # Used fixed price instead of interval for realism.
    'Ticket price - 280$\n'
    'Number of transfers - ' + transfers + ' or more\n'
    'Refund - ' + refund + '\n'
    'Luggage - ' + luggage + '\n'
    )
else:
    # Transforming price for this "special" category.
    if int(price) > 250:
        price = 'more than 250$'
    elif 200 <= int(price) <= 250:
        price = 'between 200$ and 250$'
    else:
        price = 'less than 250$'
        
    print(
    'The offer for Moscow - New York flight:\n'
    'Ticket category - Other\n'
    'Ticket price - ' + price + '\n'
    'Number of transfers - ' + transfers + '\n'
    'Refund - ' + refund + '\n'
    'Luggage - ' + luggage + '\n'
    )
```
