## Final game

<u>Short description:</u>

The game's name is **"City 404"**.  
It is about a man, who woke up in the middle of the street in unknown city. He doesn't remember how he got there. He is  exploring the area, trying to survive and get home.  
During the game player meets characters (NPC), monsters and different events. The game consists of several locations. The player choose where to move by entering letters. What events will happen depends on his choice. During exploring the area, player finds food, weapon, ammunition.
    
<u>The main goal - to survive and get back home</u>:  

   * To find a car and a gas, and then leave the city  
    
<u>The game will end if the character dies</u>. 

   * It can happen when HP becomes 0:  



### Implementing

    To make such a game we will import different libraries.
    
    1. Starting the game
    The starting menu will allow us to start the game, to read the rules of the game, to reset game, or to set different options. Menu should be available for using during the whole game, except fights. Functions to input and store  
    characters statistics (name, sex, age) will be made in the starting locations. Game logo will be shawn here also.
    
    2. Dialogs
    Funtion for dialogs will have some delay for printing, but the player can disable it in the menu, if it is annoying.
    
    3. Statistics and counting
    The statistics of characters, monsters will be stored in dictionaries. Items, which will be collected during the game, will also be stored in dictionaries.
    Functions to make action and count parameters will take the values from dictionaries and set new to them.
    
    4. Moving
    For moving our character we will make a list of acceptable values of coordinates. Function will count the coordinates according to keys entering from a keyboard, and then check if they are in list to avoid unacceptable move.
    
    5. Fights
    For fight we will use random library just not to make the damage static. Funtion that counts damage during the fight   will also counts different combinations, which are stored in dictionary. The fight continues until player's or  
    monster's HP = 0
    
    6. Items
    The events will depend on items, like keys, will be in the inventory (dictionary) or not.
    Items will dissaper when the player collects it. Item's icons will also be in a dictionary. To dissapear from the map, the icon should be changed to empty for that. 
    
    7. Different events
    Function will make different events according to the placement of the character. For that purpose it will use curent   coordinates. So, when player comes to the needed place, the event will happen.
